```python
risk_columns = [
    "Security concerns (e.g., hacking, fraud)",
    "Price volatility of Decentralized payment systems (DPS) .",
    "Lack of regulatory protection.",
    "Loss of funds.",
    "Counterparty risk."
]

unique_values = {col: data[col].unique() for col in risk_columns}
unique_values


# Mapping of categorical responses to numerical values for statistical calculation
category_map = {
    "Very High": 5,
    "High": 4,
    "Neutral": 3,
    "Low": 2,
    "Very Low": 1
}
```

```python
# Recalculate the risk analysis with percentage breakdowns for each category
total_responses = data[risk_columns[0]].count()  # Assuming equal count across all risk columns

risk_stats_with_percentages = {}
for col in risk_columns:
    # Map responses to numeric values
    mapped_data = data[col].map(category_map)
    # Calculate frequency of each category
    frequencies = data[col].value_counts().reindex(["Very High", "High", "Neutral", "Low", "Very Low"]).fillna(0)
    # Calculate mean and standard deviation
    mean = mapped_data.mean()
    std_dev = mapped_data.std()
    # Calculate percentages
    percentages = (frequencies / total_responses * 100).round(2)
    # Store results
    risk_stats_with_percentages[col] = frequencies.tolist() + percentages.tolist() + [mean, std_dev]

# Convert to DataFrame for display
risk_analysis_with_percentages_df = pd.DataFrame(risk_stats_with_percentages, index=["Very High", "High", "Neutral", "Low", "Very Low", 
                                                                                      "Very High %", "High %", "Neutral %", "Low %", "Very Low %",
                                                                                      "Mean", "Standard Deviation"])
# Calculate the grand mean across all risk columns at the percentage level as well
risk_analysis_with_percentages_df.loc["Grand Mean"] = [risk_analysis_with_percentages_df.loc["Mean", col].mean() for col in risk_columns]
risk_analysis_with_percentages_df
```

```python
benefit_columns_queries = [
    "convenience and speed of transactions",
    "lower transaction fees compared to traditional methods",
    "access to financial services for the unbanked and underbanked",
    "potential for higher investment returns",
    "increased control over finances"
]

# Search for matching column names in the data
benefit_columns = {query: [col for col in data.columns if query in col.lower()] for query in benefit_columns_queries}
benefit_columns
```

```python
# Correctly access each column name from the dictionary and check unique values
benefit_columns_flat = [col[0] for col in benefit_columns.values()]  # Flatten the list of lists

# Display unique values for each benefit column
benefit_unique_values = {col: data[col].unique() for col in benefit_columns_flat}
benefit_unique_values
```

```python
# Assign numerical values to the responses for statistical calculations
response_map = {
    "Very High": 5,
    "High": 4,
    "Neutral": 3,
    "Low": 2,
    "Very Low": 1
}

# Prepare DataFrame for analysis results
benefit_stats = {}

# Process each benefit column
for col in benefit_columns_flat:
    # Map responses to numeric values
    mapped_data = data[col].map(response_map)
    # Calculate frequency and percentage for each category
    frequencies = data[col].value_counts().reindex(["Very High", "High", "Neutral", "Low", "Very Low"]).fillna(0)
    percentages = (frequencies / frequencies.sum() * 100).round(2)
    # Calculate mean and standard deviation
    mean = mapped_data.mean()
    std_dev = mapped_data.std()
    # Store results
    benefit_stats[col] = frequencies.tolist() + percentages.tolist() + [mean, std_dev]

# Convert to DataFrame
benefit_analysis_df = pd.DataFrame(benefit_stats, index=["Very High", "High", "Neutral", "Low", "Very Low",
                                                         "Very High %", "High %", "Neutral %", "Low %", "Very Low %",
                                                         "Mean", "Standard Deviation"])
# Calculate the grand mean across all benefit columns
benefit_analysis_df.loc["Grand Mean"] = [benefit_analysis_df.loc["Mean", col].mean() for col in benefit_columns_flat]
benefit_analysis_df
```

```python
# Define response mapping and remove the incompatible column
response_mapping = {
    "Strongly agree": 5,
    "Agree": 4,
    "Neutral": 3,
    "Disagree": 2,
    "Strongly disagree": 1
}

# Remove the column with incompatible scale
#dps_benefit_cols.remove('Increased control over finances.')

# Initialize dictionary to store statistics
benefits_stats = {}

# Process each column to calculate counts, percentages, means, and standard deviations
for col in dps_benefit_cols:
    mapped_values = data[col].map(response_mapping)  # Map textual responses to numerical
    counts = data[col].value_counts()  # Count occurrences of each response
    total_responses = counts.sum()
    percentages = (counts / total_responses * 100).round(2)  # Calculate percentages
    mean = mapped_values.mean()  # Calculate mean
    std_dev = mapped_values.std()  # Calculate standard deviation
    
    # Store results
    benefits_stats[col] = counts.reindex(response_mapping.keys()).fillna(0).tolist() + percentages.reindex(response_mapping.keys()).fillna(0).tolist() + [mean, std_dev]

# Convert dictionary to DataFrame for better visualization
benefits_analysis_df = pd.DataFrame(benefits_stats, index=list(response_mapping.keys()) + [f"{resp} %" for resp in response_mapping.keys()] + ["Mean", "Standard Deviation"])

# Calculate the grand mean
grand_mean = sum(benefits_stats[col][-2] for col in benefits_stats) / len(benefits_stats)
#benefits_analysis_df.loc["Grand Mean"] = [grand_mean] * (2 * len(response_mapping) + 2)  # Fill grand mean row

benefits_analysis_df
```
